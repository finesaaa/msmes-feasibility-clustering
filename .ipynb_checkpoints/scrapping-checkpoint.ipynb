```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.bi.go.id/bisaid/bisaid/GetSektor?Condition=%7C'
page = requests.get(url)
page_json = page.json()
page_json
```

```python
df_sektor = pd.DataFrame(page_json)[['SEKTOR', 'JML_PERUSAHAAN']]
print(df_sektor.shape)
df_sektor
```

```python
df_detail = pd.DataFrame()
for i in range(df_sektor.shape[0]):
    print(i+1)
    if i == 13:
        continue
    else:
        detail_url = 'https://www.bi.go.id/bisaid/bisaid/GetSektorDetail?Condition=ALL%7CALL%7C'+ str(i+1) + '%7C0%7C10'
        df_temp = pd.DataFrame(requests.get(detail_url).json())
        if i == 0:
            df_detail = df_temp
        else:
            pd.concat([df_detail, df_temp])
```

```python
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
table1 = soup.find('table', id='tb_sektor')
table1
```

```python
headers = []
for i in table1.find_all('th'):
    header = i.text
    headers.append(header)
    
headers
```

```python
df = pd.DataFrame(columns=headers)
```

```python
for i in table1.find_all('tbody'):
    print(i.text)
```

```python
for j in table1.find_all('tr')[1:]:
    row_data = j.find_all('td')
    row = [i.text for i in row_data]
    length = len(df)
    print(row)
    df.loc[length] = row
```
