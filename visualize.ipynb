## Import Libraries

```python
import pandas as pd
import os
import glob

# exploratory data analysis
import matplotlib.pyplot as plt

%matplotlib inline
```

## Load data and combine all data

```python
FILEPATH = os.path.join(os.getcwd(), "output")
list_dir = os.listdir(FILEPATH)
list_dir
```

```python
files = os.path.join(FILEPATH, "docs_details_*.csv")
files
```

```python
files = glob.glob(files)
```

```python
# combine all files in the list
combined_csv = pd.concat([pd.read_csv(f, encoding='ISO-8859-1') for f in files])
# export to csv
OUTPUT_FILE = './output/all_data_combined.csv'
combined_csv.to_csv(OUTPUT_FILE, index=False)
```
